### Problem 2


Consider a source having a trajectory, model the sound at a stationary receiver.


For the problem the receiver is a balloon at $z=1000m$ and the source is an airplane moving at $|v|=100 m/s$

After thinking about how to create a trajectory with constant velocity, I'm going to start with just a simple fly by trajectory, which should be trivial.  If I have time I may try to create a generating function that will perhaps follow a path, and generate the trajectory point by point so as to maintain a constant velocity.

Plane trajectory (source trajectory), at point $z_s = z_0$

$\vec{r}(t) = x_s \hat{x} + (y_s + |v| t) \hat{y} + z_0 \hat{z}$

$\vec{v}(t) =  |v| \hat{y} $

Balloon (receiver) trajectory at a constant location.

$\vec{r}_0(t) =  z_0 \hat{z}$

Distance vector

$|\vec{r}_s(t) - \vec{r}_0(t)| = \sqrt{ x_s^2 + |v|^2 t^2 }$

For given SPL at 1m of 120 dB re $20 \mu Pa$

$A = \sqrt{2} P_{rms}(1m)$ Units [Pa m]

Calculating doppler shift for a stationary receiver, ($\hat{r}$ is a unit vector from receiver to source) 

$f = f_0 /(1 + \frac{\vec{v_s} \cdot \hat{r}}{c} )$

**Note** for c = $340m/s$ and $v=100m/s$ this shift will be around 30% of the fundamental value (at the receiver)

Although my analysis thus far is super simple, it wouldn't be too hard to extend to arbitrary trajectories in 3 dimensions.

```python
# extremely simple trajectory - a line with CPA at x = 100m
x0=100; v=100;    # [m], [m/s]
t=np.linspace(-100,100,201)
x=np.full(201,x0)
y=t*v
dist=np.sqrt(x0**2 + v**2*t**2)
```

```python
plt.plot(x,y); plt.grid()
plt.title("Trajectory seen from above at $x=100m$ moving along $+\hat{y}$ "); plt.xlabel("x axis [m]"); plt.ylabel("y axis [m]")
plt.show()
```

```python
plt.plot(t,dist); plt.grid()
plt.title("Distance (source to receiver) vs t"); plt.xlabel("time [s]"); plt.ylabel("Distance [m]")
plt.show()
```

```python
# Plot sound pressure vs time.  (SPL 10 log(Prms^2/Pref^2))
A = np.sqrt(2)*np.sqrt((20e-6)**2 * 10**12 )

print("Sanity check value for A: SPL at 1 m: {:.3f} dB".format(10*np.log10(((A/np.sqrt(2))**2)/(20e-6)**2)))
```

```python
# very simple estimate of pressure; P ~ A/r
P_rms = (A/np.sqrt(2))/dist
SPL = 10*np.log10(P_rms**2/(20e-6)**2)
```

```python
plt.plot(t,SPL); plt.grid()
plt.title("SPL vs t"); plt.xlabel("time [s]"); plt.ylabel("dB re $(20 \mu Pa)^2$")
plt.show()
```

```python
# compute doppler shift at receiver, can do this in vectorize manner I believe directly with numpy functions
f_0 = 120; c = 340
dist = np.reshape(np.zeros(2*t.shape[0]), (2, t.shape[0]))
vel = np.reshape(np.zeros(2*t.shape[0]), (2, t.shape[0]))
```

```python
# create distance vector and velocity vector
dist[0] = x
dist[1] = y
vel[1] = v
```

```python
t.shape[0]
```

```python
# dang it, couldn't figure out a way to vectorize the series of dot products.. using a loop
f = np.zeros(t.shape[0])
for idx in range(t.shape[0]):
    # f = f_0*(1/(1 + np.dot(v,x)/(c*mag(x)))
    vdx = np.dot(vel[:,idx],dist[:,idx])
    mag = np.sqrt(np.dot(dist[:,idx],dist[:,idx]))
    shift = vdx / (mag * c)
    f[idx] = f_0/(1 + shift)
```

```python
plt.plot(t,f); plt.grid()
plt.title("Frequency [at receiver] vs t"); plt.xlabel("time [s]"); plt.ylabel("Hz")
plt.show()
```
