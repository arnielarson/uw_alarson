```python
CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 8000
RECORD_SECONDS=1

s = p.open(format=FORMAT,
          channels=CHANNELS,
          rate=RATE,
          input=True,
          frames_per_buffer=CHUNK)
print("* recording")

nframes = 0
frames = []

for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
    data = s.read(CHUNK)
    nframes+=1
    frames.append(data)
      
print("* done recording {} frames".format(nframes))
      
s.stop_stream()
s.close()
p.terminate()
```

```python
def plot_spectrogram(s, fs=8000, nF=200, nps=1024, title="Spectrogram"):
  print("Preparing spectrogram for plotting, (rate={})".format(fs))
  f, t, S = signal.spectrogram(s, fs) #, nfft=256*2)
  # worry about this later..
  #print("[Debug] f shape: {}, t shape: {}, S shape: {}".format(f.shape, t.shape, S.shape))
  #dF = f[1]-f[0]
  #print("[Debug] dF approximately {}".format(dF))
  plt.pcolormesh(t, f[:200], S[:200], shading="auto", norm=colors.LogNorm(vmin=S.min() + 1e-10, vmax=S.max()))
  plt.xlabel("Time [s]"), plt.ylabel("Frequency [Hz]")
  plt.title(title)
  plt.show()
    
```
