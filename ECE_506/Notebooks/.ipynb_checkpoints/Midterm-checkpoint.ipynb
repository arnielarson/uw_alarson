### Problem 1

Digital modulation

$R = 1/T = 1kHz$ is the bit rate/data rate.  Therefore $T = 1^{-3}s = 1ms $

The pulse is on a carrier with frequency $F_c = 1MHz = 1^{-6}Hz$  I assume the pulse can be equal to zero outside of the pulse range, and takes the form between 0 and T described below.

<center>A zero bit: $s_0(t) = + A cos(2 \pi f_c t) + 2 A sin(2 \pi f_c t )$ </center>

<center>A one bit: $s_1(t) = - A cos(2 \pi f_c t) - 2 A sin(2 \pi f_c t )$ </center>

- **Transmitter Block** - given a digital stream, this will be modulated onto the pulse and then transmitted.

- **Energy**  I calculated the energy by hand (See below) and find that $E = \frac{5}{2}A^2 T \rightarrow A  = \sqrt{\frac{2E}{5T}}$

- **Receiver Block** - including addition of noise, the carrier signal is first correlated with $s_0(t)$ adn then passed to a decision process.  

- **What is the bandwidth?**  I had some difficulty with this.  I ultimately computed $S(f)$ which has multiple terms and is essentially dominated by a term with $\frac{e^{-2\pi(f-f_c)T}}{4\pi(f-f_c)}$  I then plotted the spectrum $|S(f)|$ and found the null to null bandwidth to be about $2kHz$, or $2/T$.  Plotting and full spectrum calculation below.

```python
# pulse R = 1/T = 1e3 Hz.  T = 1e-3
# For problem, 1000 carrier cycles per T - it's somewhat hard to visualize what's happening
T=1e-3; f = 1e6; N=10000;   # Note - doubling the signal strength quadruples the max value of Rss
DT=T/N
A=1
t = np.linspace(0, T, N)
# Digital 0
s0 = A*np.cos(2*np.pi*f*t)+2*A*np.sin(2*np.pi*f*t)
# Digital 1
s1 = -A*np.cos(2*np.pi*f*t)-2*A*np.sin(2*np.pi*f*t)
a = signal.correlate(s0,s0, mode='full', method='direct')*DT
b = signal.correlate(s1,s0, mode='full', method='direct')*DT
t2 = np.linspace(0, 2*T, 2*N-1)

fig, ax = plt.subplots(2, 2, figsize=(10,8))
ax[0,0].plot(t, s0); ax1.grid()
ax[0,0].set_title("Signal s0")
ax[0,1].plot(t2, a); ax2.grid()
ax[0,1].set_title("Rss(t) ~ Autocorrelation s(t)")
ax[1,0].plot(t, s1); ax1.grid()
ax[1,0].set_title("Signal s1")
ax[1,1].plot(t2, b); ax2.grid()
ax[1,1].set_title("Rss(t) ~ Autocorrelation s(t)")
```

```python
## So the autocorrelation is maximized at T = 10e-3, let's zoom in on that
## Note N = 10000 corresponds to point T
## In this you can at least see that at T, the autocorrelation value is at + or - it's max
fig, ax = plt.subplots(2, 2, figsize=(10,8))
ax[0,0].plot(t[9900:], s0[9900:]); ax1.grid()
ax[0,0].set_title("Signal s0")
ax[0,1].plot(t2[9900:10100], a[9900:10100]); ax2.grid()
ax[0,1].set_title("Rss(t) ~ Autocorrelation s(t)")
ax[1,0].plot(t[9900:], s1[9900:]); ax1.grid()
ax[1,0].set_title("Signal s1")
ax[1,1].plot(t2[9900:10100], b[9900:10100]); ax2.grid()
ax[1,1].set_title("Rss(t) ~ Autocorrelation s(t)")
```

```python
# Spectrum, fc = 10e6.  T = 10e-3 (1MHz and 1ms)
# plot |S(f)| on log scale
A=1; T=1e-3; fc=1e6+0.001
f = np.linspace(9.95e5,1.01e6, 10000)
s= A*(4-1j)*(np.exp(-1j*2*np.pi*(f-fc)*T)-1)/(4*np.pi*(f-fc)) +\
   A*(1j-4)*(np.exp(-1j*2*np.pi*(f+fc)*T)-1)/(4*np.pi*(f+fc))
   

spectrum = s*np.conj(s)
plt.plot(f,spectrum)
```
