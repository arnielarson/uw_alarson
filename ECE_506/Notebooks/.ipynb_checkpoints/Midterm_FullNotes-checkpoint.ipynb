### Problem 2

An analog baseband source is bandlimited to $20kHz$ and has duration $100 s$.  Sampling is done at the Nyquist rate for $20kHz$ with 16 bits collected per sample, (obviously an audio signal).  
At passband consider transmission with BPSK (M=2) and QPSK (M=4) and having 50% excess bandwidth, the transmission is to have a duration of 10 seconds.

**What is the Nyquist sampling rate required to sample the source?**  The sample period will be such that $1/2T_{nyquist} = f_{source}$, which leads to a sampling frequency of twice source.  $f_{nyquist} = 2f_{source} = 40 kHz$.  

**How large is the payload?**  This leads to a payload with $100s  \times 16 bits \times f_{nyquist} = 64 \times 10^6 bits$

**Describe BPSK transmission** BPSK has 1 dimension with 1 bit per symbol.  To send data at a rate of $6.4 \times 10^6$ bits / second requires sending $6.4 \times 10^6$ symbols each second.    At pass band, and including excess bandwidth, this corresponds to a minimum bandwidth $BW = 150\% \times 2 / T$.  This bandwidth is $19.2 MHz$ for BPSK and gives a spectral bandwidth of about 1/3 bit/sec/Hz.

**Describe QPSK transmission** QPSK has 2 dimensions and sends 2 bits per symbol.  To send data at a rate of $6.4 \times 10^6$ bits / second requires sending $3.2 \times 10^6$ symbols each second.  At pass band, and including excess bandwidth, this corresponds to a minimum bandwidth $BW = 150\% \times 2 / T$.  This is $9.6 MHz$ for QPSK and gives a spectral bandwidth of about 2/3 bit/sec/Hz.


```python
f=20e3    # This is the desired frequency captured
T=1/(2*f)    # Nyquist criteria, minimal sample rate to sample this frequency, f=1/2T
print("sampling T = {}".format(T))
fs=1/T
print("Nyquist sampling fs {}".format(fs))
print("Over 100 seconds with 16 bits/ sample, the payload has {} mega bits Mb".format(100*16*fs*1e-6))
# sending 64 Mbits in 10 seconds, 6.4M symbols sec at 1 bit per symbol
print("BPSK, bandwidth = 1/T")
```

```python
# Bandwidth at passband.  Will be 2/T * excess bandwidth.
#Minimal bandwidth = 2/T
MB=2*6.4*1e6
#Required transmission BW
RMB = 1.5*MB
print("Required transmission bandwidth for BPSK is {} MHz".format(RMB*1e-6))
print("Yields Spectral efficiency {:.2} bits/sec/Hz".format(6.4e6/RMB))
# For QPSK - we send half as many symbols per second
QRMB = 2*3.2*1e6 * 1.5
print("Required transmission bandwidth for QPSK is {} MHz".format(QRMB*1e-6))
print("Yields Spectral efficiency {:.2} bits/sec/Hz".format(6.4e6/QRMB))
```

### Appendix A - Simulation and other code for Problem 1

```python
# Simulations of binary decision process in problem 1
# sigma = sqrt(No/2), Eb/No = Er = Energy Ratio
# a = sqrt(Eb)
Er=np.arange(1,12,0.5)
logEr=10*np.log10(Er)
Pe=[]
sigma = np.sqrt(1/2); N=5e5;  
rng = np.random.default_rng()
for er in Er:
    a=np.sqrt(er); 
    S1 = rng.normal(a, sigma, int(N))   # error when RV trial S1 < 0
    S2 = rng.normal(-a, sigma, int(N))  # error when RV tiral S2 > 0
    errors_s1 = sum(S1<0)
    errors_s2 = sum(S2>0)
    error_rate = (errors_s1 + errors_s2)/(2*N)
    Pe.append(error_rate)    
```

```python
# Calculate Pe with Q function; Q(x) = 0.5 * erfc(x/sqrt(2))
Pe_theoretical = 0.5*special.erfc(np.sqrt(Er))
```

```python
# Value of A and E when P_e = 1e-3
E7 = 10**(6.79/10)
print("Value of E when Pe = 1e-3 is {:.4f}No".format(E7))
A = np.sqrt(2*E7/(5*1e-3))
print("Value of A for Pe = 1e-3 is {:.4f}".format(A))
```

```python
# putting plot here so I can embed it better in main text, expects Pe and logEr
def plotsim():
    fig, ax = plt.subplots()
    ax.semilogy(logEr,Pe,"*",label='Simulation')
    ax.semilogy(logEr, Pe_theoretical, 'lightgreen', label='Theoretical')
    ax.legend()
    ax.set(xlabel="$E_b/N_o$ [dB]", ylabel="$P_e$")
    ax.set(title="Error Probability $P_E$ vs Energy $E_b/N_o$ [dB]"); 
    ax.grid()
plotsim()
```

```python
# Plotting signal and autocorrelation
# pulse R = 1/T = 1e3 Hz.  T = 1e-3
# For problem, 1000 carrier cycles per T - it's somewhat hard to visualize what's happening
T=1e-3; f = 1e6; N=10000;   # Note - doubling the signal strength quadruples the max value of Rss
DT=T/N
A=2
t = np.linspace(0, T, N)
# Digital 0
s0 = A*np.cos(2*np.pi*f*t)+2*A*np.sin(2*np.pi*f*t)
# unit energy MF
h = np.cos(2*np.pi*f*t)+2*A*np.sin(2*np.pi*f*t)
# Digital 1
s1 = -A*np.cos(2*np.pi*f*t)-2*A*np.sin(2*np.pi*f*t)
a = signal.correlate(s0,h, mode='full', method='direct')*DT
b = signal.correlate(s1,h, mode='full', method='direct')*DT
t2 = np.linspace(0, 2*T, 2*N-1)

fig, ax = plt.subplots(2, 2, figsize=(10,8))
ax[0,0].plot(t, s0); ax1.grid()
ax[0,0].set_title("Signal s0")
ax[0,1].plot(t2, a); ax2.grid()
ax[0,1].set_title("Rss(t) ~ Autocorrelation s(t)")
ax[1,0].plot(t, s1); ax1.grid()
ax[1,0].set_title("Signal s1")
ax[1,1].plot(t2, b); ax2.grid()
ax[1,1].set_title("Rss(t) ~ Autocorrelation s(t)")
```

```python
## So the autocorrelation is maximized at T = 10e-3, let's zoom in on that
## Note N = 10000 corresponds to point T
## In this you can at least see that at T, the autocorrelation value is at + or - it's max
fig, ax = plt.subplots(2, 2, figsize=(10,8))
ax[0,0].plot(t[9900:], s0[9900:]); ax1.grid()
ax[0,0].set_title("Signal s0")
ax[0,1].plot(t2[9900:10100], a[9900:10100]); ax2.grid()
ax[0,1].set_title("Rss(t) ~ Autocorrelation s(t)")
ax[1,0].plot(t[9900:], s1[9900:]); ax1.grid()
ax[1,0].set_title("Signal s1")
ax[1,1].plot(t2[9900:10100], b[9900:10100]); ax2.grid()
ax[1,1].set_title("Rss(t) ~ Autocorrelation s(t)")
```

```python
# Plotting the spectrum, fc = 10e6.  T = 10e-3 (1MHz and 1ms)
# plot |S(f)| ; zeroed in on the main narrow  peak
A=1; T=1e-3; fc=1e6+.0001
f = np.linspace(9.95e5,1.01e6, 10000)
s= A*(4-1j)*(np.exp(-1j*2*np.pi*(f-fc)*T)-1)/(4*np.pi*(f-fc)) +\
   A*(1j-4)*(np.exp(-1j*2*np.pi*(f+fc)*T)-1)/(4*np.pi*(f+fc))

spectrum = s*np.conj(s)
fig, ax = plt.subplots()
ax.plot(f,np.real(spectrum)); ax.set(xlabel="Frequency [MHz]")
ax.set(title="|S(f)|")
```
