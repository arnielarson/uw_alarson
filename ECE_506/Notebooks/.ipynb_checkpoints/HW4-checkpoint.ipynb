### Problem 1

Recreate BPSK simulation and plot results.  For this case, I mostly follow what was done in the midterm.  For BPSK, $E_s = E_b$ and the SNR is $\frac{N_o}{2}E_b$.  Assuming equally likely symbols we can generate a simulation for the waterfall curve of $P_e$ vs. $10 log_{10} (E_b/N_o)$



```python
# Simulations of binary decision process in problem 1
# sigma = sqrt(No/2), Eb/No = Er = Energy Ratio
# a = sqrt(Eb)
Er=np.arange(1,9,0.5)
logEr=10*np.log10(Er); Pe_AGWN = [];
sigma = np.sqrt(1/2); N=5e4;  
rng = np.random.default_rng()
for er in Er:
    a=np.sqrt(er); 
    S1 = rng.normal(a, sigma, int(N))   # error when RV trial S1 < 0
    S2 = rng.normal(-a, sigma, int(N))  # error when RV tiral S2 > 0
    errors_s1 = sum(S1<0)
    errors_s2 = sum(S2>0)
    error_rate = (errors_s1 + errors_s2)/(2*N) 
    Pe_AGWN.append(error_rate)
```

```python
fig, ax = plt.subplots()
ax.semilogy(logEr,Pe_AGWN,"*",label='Simulation')
ax.set(xlabel="$E_b/N_o$ [dB]", ylabel="$P_e$")
ax.set(title="[BPSK Simulation] $P_E$ vs Energy $E_b/N_o$ [dB]"); 
ax.grid()
```

### Problem 2

Add channel fading to BPSK with a model that applies a multiplier $h$ to the signal.

```python
# Include Channel Fading
# Simulations of binary decision process with single channel fading effect
# sigma = sqrt(No/2), Eb/No = Er = Energy Ratio, (No = 1)
# a = h*sqrt(Eb)
Er_f1=np.arange(1,160,6)
logEr_f1=10*np.log10(Er_f1); Pe_f1 = [];
sigma = np.sqrt(1/2); N=int(5e5);  
for er in Er_f1:
    a=np.sqrt(er) 
    S1_f1 = a*np.sqrt(-np.log(rng.uniform(0,1,N))) + rng.normal(0, sigma, N)   # error when RV trial S1 < 0
    S2_f1 = -1*a*np.sqrt(-np.log(rng.uniform(0,1,N))) + rng.normal(0, sigma, N)  # error when RV trial S2 > 0   errors_s1f = sum(S1f<0)
    errors_s1f1 = sum(S1_f1<0)
    errors_s2f1 = sum(S2_f1>0)
    error_rate_f1 = (errors_s1f1 + errors_s2f1)/(2*N) 
    #print("Er: {:.2f}; Error rate (with fading): {:.2f}".format(er, error_rate_f1))
    Pe_f1.append(error_rate_f1)
```

```python
# Debug
print("{}; {}; {}; {}".format(len(Pe_AGWN), len(logEr), len(logEr_f1), len(Pe_f1)))
```

```python
# putting plot here so I can embed it better in main text, expects Pe and logEr
fig, ax = plt.subplots(1,1,figsize=(7,5))
ax.semilogy(logEr_f1,Pe_f1,".",label='Simulation with 1 Fading Channel')
ax.semilogy(logEr,Pe_AGWN,"*",label='Simulation with AGWN')
ax.legend()
ax.set(xlabel="$E_b/N_o$ [dB]", ylabel="$P_e$")
ax.set(title="[BPSK Simulation] $P_E$ vs Energy $E_b/N_o$ [dB]"); 
ax.grid()
```

### Problem 3

Simulate the effect of multiple channels, using the same channel fading model from **Problem 2**.  Here I simulate 2, 4, 8, 16 channels, and select the max $h$ to simulate diversity.  We do not dive into actual realizations of how the max $h$ is selected here;  We simply note the strong improvement of error rates with increased channel diversity under this model.

Simulations for BPSK were completed with $N=10^6$ trials.

```python
# Generate n channels (assumes N is big compared to n)
# Model is h=sqrt(-ln(u))
# each channel a row, for RV trial m there is a column of h, select max of the column for a RV trial
def getMaxH(N, n):
  v = np.sqrt(-np.log(rng.uniform(0,1,n*N)))
  c = v.reshape(n,N)
  maxH = np.amax(c, axis=0)
  return maxH

```

```python
# Include Multiple Channels (2, 4, 8, 16)
# Simulations of binary decision process with single channel fading effect
# sigma = sqrt(No/2), Eb/No = Er = Energy Ratio, (No = 1)
# a = h*sqrt(Eb)
Er_f2=np.arange(1,80,3)
Er_f4=np.arange(1,60,2)
Er_f8=np.arange(1,40,2)
Er_f16=np.arange(1,20,1)
logEr_f2=10*np.log10(Er_f2); logEr_f4=10*np.log10(Er_f4); 
logEr_f8=10*np.log10(Er_f8); logEr_f16=10*np.log10(Er_f16); 
Pe_f2 = []; Pe_f4=[]; Pe_f8 = []; Pe_f16 = [];
sigma = np.sqrt(1/2); N=int(5e5);  
for er in Er_f2:
    a=np.sqrt(er) 
    S1_f2 = a*getMaxH(N, 2) + rng.normal(0, sigma, N)   # error when RV trial S1 < 0
    S2_f2 = -1*a*getMaxH(N, 2) + rng.normal(0, sigma, N)  # error when RV trial S2 > 0   errors_s1f = sum(S1f<0)
    errors_s1f2 = sum(S1_f2<0)
    errors_s2f2 = sum(S2_f2>0)
    error_rate_f2 = (errors_s1f2 + errors_s2f2)/(2*N) 
    Pe_f2.append(error_rate_f2)
for er in Er_f4:
    a=np.sqrt(er) 
    S1_f4 = a*getMaxH(N, 4) + rng.normal(0, sigma, N)   # error when RV trial S1 < 0
    S2_f4 = -1*a*getMaxH(N, 4) + rng.normal(0, sigma, N)  # error when RV trial S2 > 0   errors_s1f = sum(S1f<0)
    errors_s1f4 = sum(S1_f4<0)
    errors_s2f4 = sum(S2_f4>0)
    error_rate_f4 = (errors_s1f4 + errors_s2f4)/(2*N) 
    Pe_f4.append(error_rate_f4)
for er in Er_f8:
    a=np.sqrt(er) 
    S1_f8 = a*getMaxH(N, 8) + rng.normal(0, sigma, N)   # error when RV trial S1 < 0
    S2_f8 = -1*a*getMaxH(N, 8) + rng.normal(0, sigma, N)  # error when RV trial S2 > 0   errors_s1f = sum(S1f<0)
    errors_s1f8 = sum(S1_f8<0)
    errors_s2f8 = sum(S2_f8>0)
    error_rate_f8 = (errors_s1f8 + errors_s2f8)/(2*N)     
    Pe_f8.append(error_rate_f8)
for er in Er_f16:
    a=np.sqrt(er) 
    S1_f16 = a*getMaxH(N, 16) + rng.normal(0, sigma, N)   # error when RV trial S1 < 0
    S2_f16 = -1*a*getMaxH(N, 16) + rng.normal(0, sigma, N)  # error when RV trial S2 > 0   errors_s1f = sum(S1f<0)
    errors_s1f16 = sum(S1_f16<0)
    errors_s2f16 = sum(S2_f16>0)
    error_rate_f16 = (errors_s1f16 + errors_s2f16)/(2*N)     
    Pe_f16.append(error_rate_f16)
```

```python
# Note for 8 and 16 channel - quickly collapses to zero errors which doesn't show up on plot
fig, ax = plt.subplots(1,1,figsize=(12,12))
ax.semilogy(logEr_f1,Pe_f1,"o",label='Simulation with 1 Fading Channel')
ax.semilogy(logEr_f2,Pe_f2,"+",label='Simulation with 2 Fading Channels')
ax.semilogy(logEr_f4,Pe_f4,"x",label='Simulation with 4 Fading Channels')
ax.semilogy(logEr_f8,Pe_f8,"h",label='Simulation with 8 Fading Channels')
ax.semilogy(logEr_f16,Pe_f16,"h",label='Simulation with 16 Fading Channels')
ax.semilogy(logEr,Pe_AGWN,"*",label='Simulation with AGWN')
ax.legend()
ax.set(xlabel="$E_b/N_o$ [dB]", ylabel="$P_e$")
ax.set(title="[BPSK Simulation with Fading Channels] $P_E$ vs Energy $E_b/N_o$ [dB]"); 
ax.grid()
```
