```python
# Matched Filter Theory (p372, Fundamentals of Comms Theory)
# triangle pulse
B=1; Ho = 2; N=64; 
DT=1/N  # Delta T, also N is the frequency
t0=0; t1=1

t = np.linspace(t0, t1, (t1-t0)*N)
t2 = np.linspace(t0, 2*t1, 2*(t1-t0)*N-1)
s = np.zeros(N)
for v in range(N):
    s[v] = v/63.0
h = np.flip(s)    
y = signal.convolve(s,h, mode='full', method='direct')*DT


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(8,4))
ax1.plot(t, s)
ax1.set_title("s(t), F=64samples/T")
ax2.plot(t, h)
ax2.set_title("h(t)")
ax3.plot(t2, y)
ax3.set_title("y(t)")
```
